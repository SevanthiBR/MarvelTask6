# Task 6: The Matrix Puzzle â€” Decode with NumPy & Reveal the Image

```python
import numpy as np
import matplotlib.pyplot as plt

matrix = np.load("encoded_array.npy")   

print("raw shape:", matrix.shape)
print("total elements:", matrix.size) 
```

```python
plt.imshow(matrix, cmap="gray")
plt.axis("off")
plt.show()
```

```python
n = int(np.sqrt(matrix.size)) 
print("n =", n)
assert n * n == matrix.size, "Not a perfect square â€” check file."

decoded = matrix.reshape(n, n)  
print("decoded shape:", decoded.shape)
```

```python
plt.imshow(decoded, cmap="gray")
plt.axis("off")
plt.show()
```

```python
def show(img, title=""):
    plt.imshow(img, cmap="gray")
    plt.title(title)
    plt.axis("off")
    plt.show()
```

```python
show(decoded, "reshaped (default)")
show(decoded.T, "transpose")
show(np.fliplr(decoded), "flip left-right")
show(np.flipud(decoded), "flip up-down")
show(np.rot90(decoded, k=1), "rot90 k=1")
show(np.rot90(decoded, k=2), "rot180")
show(np.rot90(decoded, k=3), "rot270")
```

```python
final_img = decoded.T   
plt.imshow(final_img, cmap="gray")
plt.title("Final Decoded Image")
plt.axis("off")
plt.show()
```

```python
plt.imsave("final_decoded.png", final_img, cmap="gray")
```

### Final Result

The scrambled matrix was successfully decoded.  
After trying various transformations (transpose, flips, rotations),  
the **Transpose** view revealed the final decoded image: a clear smiley face.
